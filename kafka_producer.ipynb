```python
from json import dumps
from kafka import KafkaProducer
import pandas as pd
import time
```

```python
producer = KafkaProducer(bootstrap_servers=['localhost:9092'],
                         value_serializer=lambda x: 
                         dumps(x).encode('utf-8'))
```

```python
item = pd.read_csv('item.csv')
voucher = pd.read_csv('voucher.csv')
```

```python
for index, i in enumerate (item['scraped_start_date'].unique().tolist()):
    data_json = eval(item[item['scraped_start_date']==i].to_json(orient='records', lines=False).replace('null','None').replace('false','False').replace('true','True'))
#     print(data_json)
    for data in data_json:
        producer.send('item', value=data)
        
        
    data_json = eval(voucher[voucher['scraped_start_date']==i].to_json(orient='records', lines=False).replace('null','None').replace('false','False').replace('true','True'))
    for data in data_json:
        producer.send('voucher', value=data)
        
    print('finish batch {}'.format(index))
    time.sleep(1)
    break
```
